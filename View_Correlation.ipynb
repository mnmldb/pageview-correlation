### import modules

```python
import pandas as pd
import numpy as np
import matplotlib.pyplot as plt
import seaborn as sns
```

### data processing

```python
df_women = pd.read_csv('cor_women.csv', dtype='float')
df_men = pd.read_csv('cor_men.csv', dtype='float')
```

```python
df_women.head(10)
```

```python
df_men.head(10)
```

```python
#replace null with 0
df_women = df_women.fillna(0)
df_men = df_men.fillna(0)
```

```python
#extract column names
brands_women = [i for i in df_women.columns if i not in ['user', 'total']]
brands_men = [i for i in df_men.columns if i not in ['user', 'total']]
```

```python
#extract only valid users
df_women_base = df_women[(df_women['total'] < 10000) & (df_women['total'] > 100)]
df_men_base = df_men[(df_men['total'] < 10000) & (df_men['total'] > 100)]
```

```python
#replace all values to either 0/1
df_women_analysis = df_women_base[brands_women].where(df_women_base[brands_women] == 0, 1)
df_men_analysis = df_men_base[brands_men].where(df_men_base[brands_men] == 0, 1)
```

### visualize

```python
#correlation matrix - women
colormap = plt.cm.Reds
corr = df_women_analysis.corr()

mask = np.zeros_like(corr, dtype=np.bool)
mask[np.triu_indices_from(mask)] = True

plt.figure(figsize=(14,12))
plt.title('Pearson Correlation of Brands (Women)', y=1.05, size=15)
sns.heatmap(corr, linewidth=0.1, vmax=1.0, square=True, cmap=colormap, linecolor='white', annot=True, mask=mask)
```

```python
#clustermap - women
colormap = plt.cm.Reds
corr = df_women_analysis.corr()

sns.clustermap(corr, cmap=colormap)
```

```python
#correlation matrix - men
colormap = plt.cm.Blues
corr = df_men_analysis.corr()

mask = np.zeros_like(corr, dtype=np.bool)
mask[np.triu_indices_from(mask)] = True

plt.figure(figsize=(14,12))
plt.title('Pearson Correlation of Brands (Men)', y=1.05, size=15)
sns.heatmap(corr, linewidth=0.1, vmax=1.0, square=True, cmap=colormap, linecolor='white', annot=True, mask=mask)
```

```python
#clustermap - men
colormap = plt.cm.Blues
corr = df_men_analysis.corr()

sns.clustermap(corr, cmap=colormap)
```
